```python
from datetime import datetime

class Vital:
    # Initializes a single vital, records the name, value, and time when it was entered
    def __init__(self, vital_name, value):
        self.__vital_name = vital_name
        self.__value = value
        self.timestamp = datetime.now()

    # Updates vital value and adds a new timestamp
    def update_value(self, new_value):
        self.__value = new_value
        self.timestamp = datetime.now()

    # Returns vital name, value, and when it was entered
    def display(self):
        formatted_time = self.timestamp.strftime("%Y-%m-%d %H:%M:%S")
        return f"{self.__vital_name}: {self.__value} | Entered {formatted_time}"

class VitalsManager:

    # Creates a dict to store all of the patient specific vitals
    def __init__(self):
        self.patient_vitals = {}

    # This is for adding vitals to a patient 
    def add_vital(self, role):
        
        # This checks the perms of the user
        if role.strip().lower() in ['doctor', 'nurse']: # I still need a way to actually check the roles of the input and not let anyone access it
            patient_id = input(f'Please enter patient id: ').strip()

            # Prevents user from leaving patient id blank
            while patient_id == "":
                print("Patient ID cannot be blank.")
                patient_id = input('Please enter patient id: ').strip()

            # Creates an emtpy vital list for patient if they dont already have one
            if patient_id not in self.patient_vitals:
                self.patient_vitals[patient_id] = []

            # Clears old vitals so the patient only has one set of current vitals
            self.patient_vitals[patient_id] = []
                
            vital_names = ['height', 'weight', 'bp', 'pulse', 'temperature']

            # Goes through all the vitals in this list^^ and asks the user to enter a value
            for vital in vital_names:
                value = input(f'Please enter {vital}: ').strip()

                # Prevents user from leaving vital value blank
                while value == "":
                    print("Value cannot be blank.")
                    value = input(f'Please enter {vital}: ').strip()
                    
                new_vital = Vital(vital, value)
                self.patient_vitals[patient_id].append(new_vital)

            print('Vitals added.')
        else:
            print(f'You are not authorized to add vitals.')

    # View vitals function to allow people to view patients vitals
    def view_vitals(self, role):
        
        # Checks the perms of the user
        if role.strip().lower() in ['doctor', 'nurse', 'patient']: # Same here^^^
            patient_id = input("Please enter patient id: ").strip()

            # Prevents user from leaving patient id blank
            while patient_id == "":
                print("Patient ID cannot be blank.")
                patient_id = input('Please enter patient id: ').strip()

            # If the patient has no vitals, this returns a message to the user
            if patient_id not in self.patient_vitals or len(self.patient_vitals[patient_id]) == 0:
                print("No vitals recorded for this patient.")
                return

            else:
                print(f"\nVitals for patient {patient_id}:")

                # Goes through each vital in the patients list and uses the display function to show it to the user
                for vital in self.patient_vitals[patient_id]:
                    print(vital.display())

        else:
            print(f'You are not authorized to view this information.')

    # Function for updating vitals if there is a typo, or new information, etc
    def update_vitals(self, role):

        # Checks the perms of the user
        if role.strip().lower() in ['doctor', 'nurse']:
            patient_id = input("Please enter patient id: ").strip()

            # Prevents user from leaving patient id blank
            while patient_id == "":
                print("Patient ID cannot be blank.")
                patient_id = input('Please enter patient id: ').strip()

            # If the patient has no vitals this shows a message to the user
            if patient_id not in self.patient_vitals or len(self.patient_vitals[patient_id]) == 0:
                print("No vitals to update for this patient.")
                return

            print(f'\nCurrent Vitals for patient {patient_id}')

            # Prints the patients vitals attached to a different number to make it easy to choose with vital to update
            for number in range(len(self.patient_vitals[patient_id])):
                print(f'{number}: {self.patient_vitals[patient_id][number].display()}')

            # This handles input errors if the user enter a nonnumerical value
            try:
                index = int(input('Enter the number of the vital you wish to update: '))
            except ValueError:
                print('Invalid choice')
                return

            # This handles if the user enters a number outside of what's displayed
            if index < 0 or index >= len(self.patient_vitals[patient_id]):
                print('Invalid choice')
                return

            new_value = input('Please enter new value: ').strip()

            # Prevents user from leaving vital value blank
            while new_value == "":
                print("Value cannot be blank.")
                new_value = input(f'Please enter new value: ').strip()

            self.patient_vitals[patient_id][index].update_value(new_value)

            print('Vital updated successfully.')
            
        else:
            print(f'You are not authorized to edit vitals.')
                    
vitals_manager = VitalsManager()

# Very rudimentary interactive menu to help me test code

while True:
    print("\nVitals Module")
    print("1. Add Vital")
    print("2. View Vitals")
    print("3. Update Vitals")
    print("4. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        role = input("Enter role: ")
        vitals_manager.add_vital(role)

    elif choice == "2":
        role = input("Enter role: ")
        vitals_manager.view_vitals(role)
        
    elif choice == '3':
        role = input('Enter role: ')
        vitals_manager.update_vitals(role)
    
    elif choice == "4":
        print("Exiting module.")
        break
    
    else:
        print("Invalid choice.")
```
