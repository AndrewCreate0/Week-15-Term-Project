```python
class Medication:
    def __init__(self, name, dosage, route, frequency):
        self.__name = name
        self.__dosage = dosage
        self.__route = route
        self.__frequency = frequency

    def __str__(self):
        return f'{self.__name}, {self.__dosage}, {self.__route}, {self.__frequency}'

    def duplicate_med(self, other):
        return str(self).lower() == str(other).lower()

class MedicationManager:
    # No way to protect against duplicate meds entered for the same patient
    def __init__(self):
        self.patient_meds = {}
    
    def prescribe_med(self, role):
        if role.strip().lower() == 'doctor': #Still stuck on how to actually check the roles so not anyone can access it
            patient_id = input("Enter patient ID: ")
            med_name = input("Enter medication name: ")
            dosage = input("Enter dosage: ")
            
            route = input("Enter route (PO, IV, Injection): ").strip().lower()
            while route not in ['po', 'iv', 'injection']:
                print(f'Invalid route. Please only enter PO, IV, or Injection for route.')
                route = input("Enter route (PO, IV, Injection): ").strip().lower()
                
            frequency = input("Enter frequency: ")

            new_med = Medication(med_name, dosage, route, frequency)

            if patient_id not in self.patient_meds:
                self.patient_meds[patient_id] = []

            for med in self.patient_meds[patient_id]:
                if med.duplicate_med(new_med):
                    print(f'Same medication is already listed for this patient.')
                    return 
                    
            self.patient_meds[patient_id].append(new_med)
            print(f"{patient_id}'s medication list has been updated.")
        else:
            print(f'You are not authorized to prescribe medication.')

    def view_med(self, role):
        if role.strip().lower() in ['doctor', 'nurse', 'patient']: # Same here ^^
            patient_id = input(f'Please enter patient id: ')
            if patient_id not in self.patient_meds or len(self.patient_meds[patient_id]) == 0:
                print(f'No medication found for this patient.')
                return
            print('\nMedications:')
            for med in self.patient_meds[patient_id]:
                print(med)
        else:
            print(f'You are not authorized to view this information.')
            
    def update_med(self, role):
        if role.strip().lower() == 'doctor':
            patient_id = input(f'Please enter the id of the patient you want to update meds for: ')
            if patient_id not in self.patient_meds or len(self.patient_meds[patient_id]) == 0:
                print(f'No medication found for this patient.')
                return
            else:
                print('\nMedications:')
                for number in range(len(self.patient_meds[patient_id])):
                    print(f"{number}: {self.patient_meds[patient_id][number]}")

            try:
                index = int(input(f'Please enter the number of the medication to update: '))
            except ValueError:
                print("Invalid choice.")
                return

            if index < 0 or index >= len(self.patient_meds[patient_id]):
                print("Invalid choice.")
                return
                
            print("\nEnter new medication details:")
            new_med_name = input("Enter medication name: ")
            new_dosage = input("Enter dosage: ")
            
            new_route = input("Enter route (PO, IV, Injection): ").strip().lower()
            while new_route not in ['po', 'iv', 'injection']:
                print(f'Invalid route. Please only enter PO, IV, or Injection for route.')
                new_route = input("Enter route (PO, IV, Injection): ").strip().lower()
                
            new_frequency = input("Enter frequency: ")

            new_med = Medication(new_med_name, new_dosage, new_route, new_frequency)
            
            for i in range(len(self.patient_meds[patient_id])):
                if i != index and self.patient_meds[patient_id][i].duplicate_med(new_med):
                    print("This medication is already listed for this patient.")
                    return

            self.patient_meds[patient_id][index] = new_med
            print("Medication updated.")
        
        else: 
            print(f'You are not authorized to update medications.')

    def delete_med(self, role):
        if role.strip().lower() == 'doctor':
            patient_id = input(f'Please enter the id of the patient you want to delete meds for: ')
            if patient_id not in self.patient_meds or len(self.patient_meds[patient_id]) == 0:
                print(f'No medication found for this patient.')
                return
            else:
                print(f'\nMedications:')
                for number in range(len(self.patient_meds[patient_id])):
                    print(f' {number}: {self.patient_meds[patient_id][number]}')
                    
                try:
                    index = int(input(f'Please enter the number of the medication you wish to delete: '))
                except ValueError:
                    print(f'Invalid choice.')
                    return

                if index < 0 or index >= len(self.patient_meds[patient_id]):
                    print(f'Invalid choice.')
                    return

                deleted_med = self.patient_meds[patient_id].pop(index)
                print(f'Deleted: {deleted_med}')

                if len(self.patient_meds[patient_id]) == 0:
                    del self.patient_meds[patient_id]
                                
        else: 
            print(f'You are not authorized to delete medications.')
            
med_manager = MedicationManager()

# The same rudimentary interactive menu to help me test the code

while True:
    print("\nMedication Module")
    print("1. Prescribe Medication")
    print("2. View Patient Medications")
    print("3. Update Patient Medication")
    print("4. Delete Medication")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        role = input("Enter role: ")
        med_manager.prescribe_med(role)

    elif choice == "2":
        role = input("Enter role: ")
        med_manager.view_med(role)

    elif choice == "3":
        role = input("Enter role: ")
        med_manager.update_med(role)

    elif choice == "4":
        role = input("Enter role: ")
        med_manager.delete_med(role)

    elif choice == "5":
        print("Exiting module.")
        break

    else:
        print("Invalid choice.")
```
