```python
from datetime import datetime

class Vital:
    def __init__(self, vitals, value):
        self.__vitals = vitals 
        self.__value = value
        self.timestamp = datetime.now() # When the time is displayed, there are like 6 decimals and idk how to remove them

    def display(self):
        print(f'{self.__vitals}: {self.__value} | {self.timestamp}')

class VitalsManager:
    # Vitals are no tied to specific patients yet and I need to fix that
    def __init__(self):
        self.vitals_list = []

    def add_vital(self, role):
        if role.lower() == 'doctor' or 'nurse': # I still need a way to actually check the roles of the input and not let anyone access it
            vital = input(f'Please enter the vital name: ')
            value = input(f'Please enter value: ')
            # User can input anything for values and names (eg: banana or Z1Gdasd%), need to figure out ways to not have that
            
            v = Vital(vital, value)
            self.vitals_list.append(v)

            print('Vital added.')
        else:
            print(f'You are not authorized to add vitals.')

    def view_vitals(self, role):
        if role == 'doctor' or 'nurse' or 'patient': # Same here
            if len(self.vitals_list) == 0:
                print(f'No vitals recorded.')
            else:
                for vital in self.vitals_list:
                    vital.display()
                    
vitals_manager = VitalsManager()

# Very rudimentary interactive menu to help me test code

while True:
    print("\nVitals Module")
    print("1. Add Vital")
    print("2. View Vitals")
    print("3. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        role = input("Enter role: ")
        vitals_manager.add_vital(role)

    elif choice == "2":
        role = input("Enter role: ")
        vitals_manager.view_vitals(role)

    elif choice == "3":
        print("Exiting module.")
        break

    else:
        print("Invalid choice.")
```
