```python
# doctor must login. Default login: Doctor ID = chief, Password = 12345.
#Doctors can create new doctors (ID and password); Create patients. Create/view all patient
#order lab tests (with costs) to a patient. Assign patient to a doctor.
#Discharge patient.First name, last name, specialty, phone number

# creat person class. 

class person:
    def __inti__(self, first_name, last_name):
        self.first_name = first_name 
        self.last_name = last_name 


def exit_HIS():
    print("thanks you for visting the HIS!")


# doctor calss. 
class Doctor:
    def __init__(self, first_name, last_name, specialty, phone, doctor_id, password):
        self.first_name = first_name 
        self.last_name = last_name 
        self.specialt = specialty 
        self.phone = phone 
        self.doctor_id = doctor_id 
        self.password = password 

class Patient:
     def __init__(self, first_name, last_name, patient_id, age, admit_date, med_costs ):
        self.first_name = first_name 
        self.last_name = last_name 
        self.patient_id = patient_id
        self.age = age 
        self.admit_date = admit_date
        self.med_costs = med_costs
        self.lab_tests = []
        
doctors = [
    Doctor("Rafa", "Younis", "Cardiology", "6025551234", "Doc101", "pass101"),
    Doctor("Sarah", "Smith", " Pediatrics", "4805552314", "Doc102", "pass102"),
    Doctor("Micheal", "lee", "Neurology", "6236668899", "Doc103", "pass103"),
    Doctor("Emily", "Davis", "Dermatology", "5206669934", "doc104", "pass104")
]
defult_doctor = Doctor("Chief", "Doctor", "General", "6027277546", "cheif", "12345")
doctors.append(defult_doctor)


patients = [
    Patient("John", "Smith", 1, 65, "March 3", 5),
    Patient("Patrica", "Jones", 2, 73, "February 9", 1000000000),
    Patient("Someone", "Something", 3, 10, "April 7", 30),
    Patient("Zack", "Xylophone", 4, 45, "December 31", 990)
]

# This function checks doctor id and password
# If doctor correct allow access
# If wrong show invalid message
def login ():
    doc_id = input("Enter Doctor ID:").strip()
    password = input("Enter password:").strip()
    # loop through all doctors in the system 
    for doctor in doctors:
        if doctor.doctor_id == doc_id and doctor.password == password:
            print("login sucessful.welcome to the HIS")
            return True  # Allow access if correct
    # IF no match access fails    
    print("invalid login.")
    return False 

#Create new_doctor function.
# This creates a new doctor and adds doctor to list
def create_doctor():
    # Get doctor inforamtion from the user
    first_name = input("Enter first name:")
    last_name = input("Enter last name:")
    specialty = input("Enter specialty:")
    phone = input(" Enter phone number:")
    doctor_id = input("Enter doctor id:")
    password = input("Enter password:")

    # Create new doctor object, add doctor to list
    new_doctor = Doctor(first_name , last_name, specialty, phone, doctor_id, password)
    doctors.append(new_doctor)
    print("doctor created successfully.")
 
     
# Create patient function. 
# This function creates a new patient and add patient to list
def create_patient():
    first_name = input(" Enter patient first name:")
    last_name = input(" Enter patient last name:")
    patient_id = input(" Enter patient id:") 
    patient_age = input("Enter patient age:")
    patient_admit_date = input("Enter patient admit date:")
    patient_med_costs = input("Enter patient med costs:")

    new_patient = Patient(first_name, last_name, patient_id, patient_age, patient_admit_date, patient_med_costs)
    patients.append(new_patient)
    print("patient created successfully.")

#Order lab test function.
# This function finds a patient and adds alab test
# It also upadtes the taotale medaical cost
def order_lab_test():
    patient_name = input(" Enter patient first name:")
    # loop through all patients to find the correct one 
    for patient in patients:
        # Check if patient match  
        if patient. first_name == patient_name:
            test_name = input("Enter lab test name:")
            cost = float(input("Enter test cost:"))
                
            patient.lab_tests.append((test_name, cost))
            # Add lab test cost to patient total cost
            patient.med_costs += cost
            
            print("Lab test ordered  sucessfully.")
            return # stop function after finding patient
            
    print("patient not found.") # If no patient was found 

# View patients function:
# This function displays all patients detalis (name, age, ect,)
def view_all_patients():
    if not patients:
        print("No patient found.") # print message if no patient exists
        return

    # Print header for patient list    
    print("\===== all patients detalies =====")
     
    # loop through each patient in patient list 
    for patient in patients:
            print("\n---------")
            print("First Name:", patient.first_name)
            print("Last Name:", patient.last_name)
            print("Patient ID:", patient.patient_id)
            print("Age:", patient. age)
            print("Admit Date:", patient.admit_date)
            print("Medicale Cost:", patient.med_costs)

    print("/n==============") # Print closing line after listing all lines

#Discharge patient.
# This function removes patients from the system. 
def discharge_patient():
    # Ask user to enter patient first name 
    patient_name = input("enter patient first name:")
    # loop through all patients in list 
    for patient in patients:
        if patient. first_name. lower()== patient_name.lower():
            patient.discharged = True # mark patientas discharged 
            print("patient discharged successfully.") # print confirmation message 
            return
            print("patient not found")
            return False 

            
# main program 
# check login before showing menu
if login():
    # keep program running untile user exists
    while True:
        # Disply menu opations
         print("\nDoctor Module")
         print("1. creat newe doctor")
         print("2. creat patient")
         print("3. view all patients")
         print("4.order lab test")
         print("5.discharge patient")
         print("6.exit")
        # Get user choice
         choice  = input("enter your choice:")
         # if user select opation 1 create doctor
         if choice == "1":
            create_doctor()
         elif choice == "2":
             create_patient()
         elif choice == "3":
            view_all_patients()
         elif choice == "4":
              order_lab_test()
         elif choice == "5":
             discharge_patient()
         elif choice == "6":
             print(" Thak you for Visiting the  HIS")
             break
         else: 
             print("invalid choise.")



    

        


    
    
    




     
     


            




        
        
    
```
