```python
class Medication:
    def __init__(self, name, dosage, route, frequency):
        self.__name = name
        self.__dosage = dosage
        self.__route = route
        self.__frequency = frequency

    def __str__(self):
        return f'{self.__name}, {self.__dosage}, {self.__route}, {self.__frequency}'

class MedicationManager:
    # I need to add update and delete med functions
    # No way to protect against duplicate meds entered for the same patient
    def __init__(self):
        self.patient_meds = {}
    
    def perscribe_med(self, role):
        if role.lower() == 'doctor': #Still stuck on how to actually check the roles so not anyone can access it
            patient_id = input("Enter patient ID: ")
            med_name = input("Enter medication name: ")
            dosage = input("Enter dosage: ")
            route = input("Enter route (PO, IV, Injection): ")
            frequency = input("Enter frequency: ")

            new_med = Medication(med_name, dosage, route, frequency)

            if patient_id not in self.patient_meds:
                self.patient_meds[patient_id] = []
                
            self.patient_meds[patient_id].append(new_med)
            print(f"{patient_id}'s medication list has been updated.")
        else:
            print(f'You are not authorized to perscribe medication.')

    def view_med(self, role):
        if role.lower() == 'doctor' or 'nurse' or 'patient': # Same here ^^
            patient_id = input(f'Please enter patient id: ')
            if patient_id not in self.patient_meds:
                print(f'No medication found for this patient.')
                return
            print('Medications:')
            for med in self.patient_meds[patient_id]:
                print(med)
                
med_manager = MedicationManager()

# The same rudimentary interactive menu to help me test the code

while True:
    print("\nMedication Module")
    print("1. Prescribe Medication")
    print("2. View Patient Medications")
    print("3. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        role = input("Enter role: ")
        med_manager.perscribe_med(role)

    elif choice == "2":
        role = input("Enter role: ")
        med_manager.view_med(role)

    elif choice == "3":
        print("Exiting module.")
        break

    else:
        print("Invalid choice.")
```
